## Data visualization for IRIS dataset using matplotlib and Seaborn


```python
# importing the required libraries
import matplotlib.pyplot as plt  
import seaborn as sns   
import pandas as pd  
import numpy as np  
```

```python
# getting the built-in datasets of seaborn because seaborn holds multiple datasets by default
print(sns.get_dataset_names())
```

```python
# loading dataset
df=sns.load_dataset("iris")
df.head()
```

```python
df["species"].unique()
#df.species.unique()
```

```python
# check for missing values
df.isnull().sum()
```

```python
df.describe().T
```

```python
df.species.value_counts()
```

```python
#Matplotlib plotting
plt.plot(df['sepal_length'],df['sepal_width'])
```

```python
sns.lineplot(x='sepal_length', y='sepal_width', data=df, errorbar=("ci",95)) #confidence interval
plt.show            
```

```python
### Line plot
sns.lineplot(x='sepal_length',y='sepal_width', data=df, hue='species', errorbar=None)
plt.show
```

##### Oberservation
- Virginica has maximum sepal_length
- Setosa has maximum sepal_width

```python
# Continuous and categorical feature
sns.lineplot(x='sepal_length', y='species', data=df, hue='species', errorbar=None)
plt.show
```

#### Observation
- Most of virginica has sepal length greater than 7 units.
- Versicolor can be found in range 5.0 units to 7.0 units.
- Setosa can be found in 4.3 to 5.5 units.
- If sepal length of flower is greater than 4.3 and less than 4.8 units than only flower can be iris(setosa) flower.

```python
sns.lineplot(x='petal_length', y='species', data=df, hue='species',errorbar=None)
plt.show
```

```python
## Scatter plot
sns.scatterplot(x='sepal_length', y='sepal_width', data=df, hue='species')
plt.title("Sepal length Vs Sepal width")
plt.show
```

```python
df.head()
```

```python
'''Bar plot (Categorical estimate plot) we use when we have one categorical data and one numerical
data.'''
sns.barplot(x='species', y='sepal_length', data=df) # Categorical data and numerical data
plt.show
```

```python
df["species"].value_counts()
```

```python
## Count Plot value_counts()
## A countplot basically counts the categorical and return a count of their occurences.
sns.countplot(x='species',data=df)
plt.show()    
```

```python
## Box Plot
sns.boxplot(x='sepal_width', data=df)
plt.show()
```

```python
sns.boxplot(x='species', y='sepal_width', data=df)
plt.show()  
## Checking outliers in each category with respect to sepal width
```

```python
## Kde (kernel density estimator) plot to view the distribution of the data
sns.kdeplot(x='sepal_width', data=df) 
plt.show()
```

```python
## violin plot (box plot and kde together)
sns.violinplot(x='sepal_width', data=df)
plt.show()
```

```python
## Strip plot

sns.stripplot(x='species', y='petal_width', data=df, hue='species')
plt.show()
```

```python
## Histogram

sns.histplot(x='petal_width', data=df, bins=[0.0, 0.5,1.0,1.5,2.0,2.5]) # bins/intervals
plt.show()
```

```python
sns.histplot(x='petal_length', data=df, kde=True)  # Kernal density estimation
plt.show()
## kde is approximaiton of histogram to density curve..
```

```python
## To hide the warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
## Dist Plot  (similar to histogram not that much used)
''' Distplot is used basically for univariant set of observations and
 visualizes it through a histogram i.e. only one oberservation and hence we choose
 one particular column of the dataset'''
sns.distplot(df['petal_width'], color="g", hist=True, kde=True)
plt.show()
## density means the probability of given point to lie on the curve.
```

```python
## Pairplot (gives the visualization for all the numerical data)
''' Pairplot represents pairwise relation across the entire dataframe and supports an
additional argument called hue for categorical separation.
What it does basically is create a joinplot between every possible numerical
column and takes a while if the dataframe is really huge'''
sns.pairplot(data=df,hue='species',diag_kind='kde')  #diag_kind=hist
plt.show()
```

```python
## Categorical plot
sns.catplot(x='sepal_length',data=df, kind='violin')
plt.show()
```

```python
# Scatter plot
sns.catplot(x='species',y='sepal_length',data=df, kind='swarm') 
plt.show()
```

```python
## The default representation of the data in cat plot() uses a stripplot
sns.catplot(x='petal_length', y='species',data=df)
plt.show()
```
